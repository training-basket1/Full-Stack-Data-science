### Concat()

```python
df6 =pd.DataFrame({"ID" : [1,2,3,4],
       "Name" : ["a", "b","c", "d"],
       "Class" : [5,6,7,8],
        "marks" :[10 , 20 , 30 , 40]})
```

```python
df6.head()
```

```python
df7 =pd.DataFrame({"ID" : [6 , 7, 8 , 9 , 10],
       "Name" : ["e", "f","g", "h" , "e"],
       "Class" : [9,10,11,12 , 7],
        "marks" : [50 , 60 , 70 , 80 , 90]})
```

```python
df7.head()
```

```python
data = pd.concat([df6,df7] , axis = 0) ### column wise
```

```python
data
```

```python
pd.concat([df7,df6] , axis = 1)
```

```python
df7["tarun"] = df7["marks"] - df7["Class"]
df7.head()
```
